```python
x = 2 
def f():
    x = 3
    return x
print(x) # x= 2 global scope
print(f()) # x = 3 local scope
```

```python
#nested function
#there is LEGB rule that is search local scope, enclosing function, global and bulit in scopes, respectivley

def square():
    def add():
        x = 2
        y = 3
        z = x+y
        return z
    return add()**2
print(square())
```

```python
# default arguments 

def f(a,b = 1, c= 2): 
    y = a+b+c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3))
```

```python
#flexible arguments *args

def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1, 2, 3, 4, 5)
```

```python
# flexible arguments **kwargs that is dictionary
def f(**Kwargs):
    """ print key and value of dictionary"""
    for key, value in Kwargs.items():
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```
